```python
from transformers import pipeline
```

```python
grammar_corrector = pipeline("text2text-generation", model="vennify/t5-base-grammar-correction")
```

```python
bad_sentence = "Me likes to go store yesterday."
corrected = grammar_corrector(bad_sentence)
```

```python
print(f"Original: {bad_sentence}")
print(f"Corrected: {corrected[0]['generated_text']}")
```
